```python
directory_path = r'/Users/boycot/EpitechProjects/ai/data'
```

```python
summarizer_T5_1 = Summarayzer(data_directory = directory_path, model_name ='t5-base' , class_name='t5') 
print(summarizer_T5_1)
```

```python
summarizer_T5_1.preprocess_text( column_names = [ 'reference_summary', 'original_text' ] )
```

```python
summarizer_T5_1.split_data(X='original_text_lower', y='reference_summary_lower', train_size=0.7, val_size=0.2, batch_size = 8)
print(summarizer_T5_1)
```

```python
summarizer_T5_1.train(epochs=3, learning_rate=0.0002, weight_decay=0.01, grad_clip=None , early_stopping_patience=None, log_interval=1, validation_interval=10)
print(summarizer_T5_1)
```
